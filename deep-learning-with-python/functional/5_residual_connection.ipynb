```python
from keras import Input
from keras.layers import Conv2D, MaxPooling2D, add
from keras.models import Model
from plot_util import show_model
```

```python
x = Input(shape=(299, 299, 128))
y = Conv2D(128, 3, activation='relu', padding='same')(x)
y = Conv2D(128, 3, activation='relu', padding='same')(y)
y = Conv2D(128, 3, activation='relu', padding='same')(y)

y = add([y, x])

model = Model(x, y)
```

```python
show_model(model)
```

```python
x = Input(shape=(299, 299, 128))
y = Conv2D(128, 3, activation='relu', padding='same')(x)
y = Conv2D(128, 3, activation='relu', padding='same')(y)
y = MaxPooling2D(2, strides=2, padding='same')(y)

residual = Conv2D(128, 1, strides=2, padding='same')(x)
y = add([y, residual])

model = Model(x, y)
```

```python
show_model(model)
```
