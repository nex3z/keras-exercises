```python
import os, shutil
```

```python
RAW_DATA_DIR = './data/dogs-vs-cats-raw'
OUTPUT_BASE_DIR = './data/dogs-vs-cats-small'
```

```python
def build_dataset(ori_data_dir, output_base_dir, name, sample_range):
    dataset_dir = os.path.join(output_base_dir, name)
    os.mkdir(dataset_dir)
    
    cats_dir = os.path.join(dataset_dir, 'cats')
    os.mkdir(cats_dir)
    cats_file_names = ['cat.{}.jpg'.format(i) for i in sample_range]
    for file_name in cats_file_names:
        shutil.copyfile(os.path.join(ori_data_dir, file_name), os.path.join(cats_dir, file_name))
    
    dogs_dir = os.path.join(dataset_dir, 'dogs')
    os.mkdir(dogs_dir)
    dogs_file_names = ['dog.{}.jpg'.format(i) for i in sample_range]
    for file_name in dogs_file_names:
        shutil.copyfile(os.path.join(ori_data_dir, file_name), os.path.join(dogs_dir, file_name))
```

```python
os.mkdir(OUTPUT_BASE_DIR)
build_dataset(os.path.join(RAW_DATA_DIR, 'train'), OUTPUT_BASE_DIR, 'train', range(1000))
build_dataset(os.path.join(RAW_DATA_DIR, 'train'), OUTPUT_BASE_DIR, 'validation', range(1000, 1500))
build_dataset(os.path.join(RAW_DATA_DIR, 'train'), OUTPUT_BASE_DIR, 'test', range(1500, 2000))
```
