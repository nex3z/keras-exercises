```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from BostonData import BostonData
import util
```

```python
data = BostonData()
```

```python
def build_model():
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(data.train_x.shape[1], )))
    model.add(Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model
```

```python
k = 4
num_epochs = 100
# num_epochs = 200
all_mae_histories = []
for i in range(k):
    print("processing fold {}".format(i))
    (train_x, train_y), (val_x, val_y) = data.get_fold(i, k)
    model = build_model()
    history = model.fit(train_x, train_y, validation_data=(val_x, val_y), 
                        epochs=num_epochs, batch_size=1, verbose=1)
    mae_history = history.history['val_mean_absolute_error']
    all_mae_histories.append(mae_history)
```

```python
history = np.array(all_mae_histories)
history_mean = history.mean(axis=0)

smoothed = util.moving_average(history_mean[10:])
plt.plot(range(1, len(smoothed) + 1), smoothed)
plt.xlabel('Epochs')
plt.ylabel('Validation MAE')
plt.show()
```

```python
model = build_model()
model.fit(data.train_x, data.train_y, epochs=80, batch_size=16, verbose=0)
result = model.evaluate(data.test_x, data.test_y)
print(result)
```
